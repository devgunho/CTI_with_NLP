```python
import modin.pandas as mpd
```

```python
df = mpd.read_csv('./dataset/dummy_input.csv', names=['CORPUS'])
df
```

```python
df['CORPUS']
```

```python
corpus = df['CORPUS'].values.tolist()
print(corpus)
print(corpus[0])
```

```python
import json
dictionary_json = dict()

for i in range(0, len(corpus)):
    key = corpus[i]

    mitre_info = [{
        "domain":
        "Enterprise",
        "_id":
        "T1071.001",
        "description":
        "admin@338 is a China-based cyber threat group. It has previously used newsworthy events as lures to deliver malware and has primarily targeted organizations involved in financial, economic, and trade policy, typically using publicly available RATs such as PoisonIvy, as well as some non-public backdoors."
    }]

    content = {}
    content["id"] = i + 1
    content["mitre"] = mitre_info
    content["pos"] = "part of speech"
    content["nltk-ner"] = "ORGANIZATION"
    content["pororo-ner"] = "PERSON"
    dictionary_json[key] = content

print(json.dumps(dictionary_json, indent='\t'))
```

```python
with open('dictionary.json', 'w', encoding="utf-8") as make_file:
    json.dump(dictionary_json, make_file, ensure_ascii=False, indent="\t")

with open('dictionary.json', 'r') as read_file:
    json_data = json.load(read_file)

print(json.dumps(json_data, indent='\t'))
```
